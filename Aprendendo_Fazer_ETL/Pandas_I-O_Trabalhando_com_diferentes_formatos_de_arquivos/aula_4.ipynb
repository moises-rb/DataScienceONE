```python
import pandas as pd
```

```python
import os

# Obtém o diretório de trabalho atual
diretorio_base = os.getcwd()

# Define o caminho relativo para o arquivo HTML
dados_html = 'filmes_wikipedia.html'
caminho_arquivo = os.path.join(diretorio_base, 'base_dados', dados_html)

# Lendo o arquivo HTML
try:
    with open(caminho_arquivo, 'r', encoding='utf-8') as f:
        conteudo_html = f.read()
except FileNotFoundError:
    print(f"Erro: O arquivo '{dados_html}' não foi encontrado no caminho: '{caminho_arquivo}'")
    exit()

# Agora a variável 'conteudo_html' contém todo o conteúdo do seu arquivo HTML como uma string.
# Você pode imprimir para visualizar (se não for muito grande).
# print(conteudo_html)

print(f"Arquivo '{dados_html}' lido com sucesso. O conteúdo está na variável 'conteudo_html'.")
```
